```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
import numpy as np
```

```python
x = load_iris().data
```

```python
print(x.shape)
```

```python
feature_names  =load_iris().feature_names
print('Feature Names : ',feature_names)
print('Labels : ',load_iris().target_names)
```

```python
x=np.array(x)
```

```python
model = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
```

```python
hc = dendrogram(linkage(x[:50,:],method='ward'))
plt.title('Hierarchy Graph')
plt.xlabel('x values')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

```python
pred_labels = model.fit_predict(x)
```

```python
print(pred_labels)
```

```python
lables = set(model.labels_)

print('Labels : ',lables)
```

```python
for i in range(3):
    plt.scatter(x[pred_labels==0,i],x[pred_labels==0,i+1], color='red',alpha=0.6,label='Cluster 0',marker='*')

    plt.scatter(x[pred_labels==1,i],x[pred_labels==1,i+1], color='green',alpha=0.6,label='Cluster 1',marker='*')
    
    plt.scatter(x[pred_labels==2,i],x[pred_labels==2,i+1], color='blue',alpha=0.6,label='Cluster 2',marker='*')
    
    plt.xlabel(feature_names[i])
    plt.ylabel(feature_names[i+1])
    plt.title(f"{feature_names[i]}"+" VS " + f"{feature_names[i+1]}")
    plt.legend()
    plt.show()
```
