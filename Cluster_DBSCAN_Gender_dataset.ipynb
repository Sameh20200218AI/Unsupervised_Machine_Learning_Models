```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_csv('Training set.csv')
```

```python
print(data.info)
```

```python
print(data.describe())
```

```python
# Checking Null values for training set
data.isnull().sum()*100/data.shape[0]
```

```python
data.shape
```

```python
fig = plt.figure(figsize=(7,9))
ax1 = fig.add_subplot(2,2,1)
ax1.boxplot(data['Height'])
ax1.set_title('Box Plot for Height')

ax2 = fig.add_subplot(2,2,2)
ax2.boxplot(data['Weight'])
ax2.set_title('Box Plot for Weight')
plt.tight_layout()
plt.show
```

```python
#remove outliers from data training
#Q1 is the 25% and Q3 is 75% of data, Q2 is 50% of data

Q1 = data.Height.quantile(0.25)
Q3 = data.Height.quantile(0.75)
IQR = Q3 - Q1
data = data[(data.Height >= Q1 - 1.5*IQR) & (data.Height <= Q3 + 1.5*IQR)] # drop rows that include outliers
```

```python
#remove outliers from data training
#Q1 is the 25% and Q3 is 75% of data, Q2 is 50% of data

Q1 = data.Weight.quantile(0.25)
Q3 = data.Weight.quantile(0.75)
IQR = Q3 - Q1
data = data[(data.Weight >= Q1 - 1.5*IQR) & (data.Weight <= Q3 + 1.5*IQR)] # drop rows that include outliers
```

```python
print(data.describe())
```

```python
#scatter row data
plt.scatter(data['Weight'],data['Height'],label='row data',marker='*')
plt.title('Data')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend()
plt.show()
```

```python
x=np.array(data[['Weight','Height']].values)
```

```python
print(x.shape)
```

```python
model = DBSCAN(eps=4,min_samples=30,metric='euclidean')
```

```python
pred_labels = model.fit_predict(x)
```

```python
print(pred_labels)
```

```python
lables = set(model.labels_)

print('Labels : ',lables)
```

```python

plt.scatter(x[pred_labels==0,0],x[pred_labels==0,1], color='red',alpha=0.4,label='Cluster 0',marker='*')
    
plt.scatter(x[pred_labels==-1,0],x[pred_labels==-1,1], color='blue',alpha=0.6,label='Cluster -1',marker='*')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight VS Height')
plt.legend()
plt.tight_layout()
plt.show()
```
